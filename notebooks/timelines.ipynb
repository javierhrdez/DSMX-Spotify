```python
music_full = pd.read_csv("data/top_200_spotify_full.csv")
```

```python
music_full.head()
```

```python
music_full.isnull().sum()
```

```python
music_total = pd.merge(music, music_full, on='url', how='left')
```

```python
music_total.shape
```

```python
music_total.dropna(inplace = True)
```

```python
music_total.isna().sum()
```

```python
#music_total.convert_objects(convert_numeric=True).head()
music_total.describe()
```

```python
#music_total.set_index('date_x').groupby(pd.Grouper(freq='D')).mean()
music_total['date_x'] = pd.to_datetime(music_total['date_x'], errors='coerce')
music_total['valence'] = music_total['valence'].astype(float)
```

```python
#music_total.resample('D', on='date_x').mean()
#music_total = music_total.convert_objects(convert_numeric=True)
music_by_day = music_total.set_index('date_x').groupby(pd.Grouper(freq='D')).mean()#.dropna()
music_by_month = music_total.set_index('date_x').groupby(pd.Grouper(freq='MS')).mean()#.dropna()
music_by_year = music_total.set_index('date_x').groupby(pd.Grouper(freq='BYS')).mean()#.dropna()

#music_total.dtypes
```

```python
#music_by_day['valence'].pct_change()
music_by_day['streams'].pct_change().sort_values(ascending = False)
```

```python
fig, ax = plt.subplots(figsize=(30,15))
#ax = sns.lineplot(data=[music_by_day.valence, music_by_day.energy, music_by_day.danceability], palette="tab10", linewidth=2.5)
#ax = sns.lineplot(data=[music_by_day.valence], palette="tab10", linewidth=2.5)
#music_by_day['valence'].pct_change()
ax = sns.lineplot(data=[music_by_day.valence], palette="tab10", linewidth=2.5)
plt.xlabel('Fecha', fontsize=20)
plt.ylabel('Positividad', fontsize=20)
plt.title('Tiempo v/s valence', fontsize=25)
plt.plot();
```

```python
fig, ax = plt.subplots(figsize=(30,15))
#ax = sns.lineplot(data=[music_by_day.valence, music_by_day.energy, music_by_day.danceability], palette="tab10", linewidth=2.5)
#ax = sns.lineplot(data=[music_by_day.valence], palette="tab10", linewidth=2.5)
#music_by_day['valence'].pct_change()
ax = sns.lineplot(data=[music_by_day.streams.pct_change()], palette="tab10", linewidth=2.5)
plt.xlabel('Fecha', fontsize=20)
plt.ylabel('Positividad', fontsize=20)
plt.title('Tiempo v/s valence', fontsize=25)
plt.plot();
```
